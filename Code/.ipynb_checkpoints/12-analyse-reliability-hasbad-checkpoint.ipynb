```python
#save_dir_temp = "../reliability-results/baseline-{i}_"
#save_dir_temp = "../reliability-resuxlts/init-orth-{i}_"
#save_dir_temp = "../reliability-results/tc-penalty-{i}_"
#save_dir_temp = "../reliability-results/02-KL-001-{i}_"
#save_dir_temp = "../reliability-results/no_init_{i}_"
save_dir_temp = "../reliability-results/good_init_{i}_"


files = [f for f in os.listdir('../reliability-results/') if all((f.startswith(save_dir_temp.split('/')[-1][0:-4]),f.endswith('.pickle')))  ]
files.sort()
print(len(files))
files
```

```python
training_logs = []
for i in range(len(files)):
    save_dir = save_dir_temp.format(i=i)
    #with open(save_dir+'training_log.pickle', 'rb') as handle:
    with open(os.path.join('../reliability-results/',files[i]), 'rb') as handle:
        data = pickle.load(handle)
    training_logs.append(data)    
print(len(training_logs))

#training_logs = [training_logs[i] for i in range(len(training_logs)) if training_logs[i]['varExps'][-1]>.97]
training_logs = [training_logs[i] for i in range(len(training_logs)) if training_logs[i]['varExps'][0]>.755]
print(len(training_logs))
```

```python
x = np.array([val['varExps'][-1] for val in training_logs])
y = np.array([val['varExps'][0] for val in training_logs])
for i in range(len(y)):
    plt.plot(y[i],x[i],'.',markersize=20)
plt.title(f'r = {np.corrcoef(x,y)[0,1]:.2f}')
plt.xlabel('VarExp 10 Epochs')
plt.ylabel('VarExp 500 Epochs')
```

```python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w
```

```python
vec = np.array([val['varExps'] for val in training_logs]).transpose()
vec[vec<0]=0
ratio = (vec[-1,:]>.8).sum()/vec.shape[1]

plt.figure(figsize=(15,4))
plt.plot(vec,alpha=.8);
plt.title(f'VAR EXP. conv. ratio {ratio:.2f}\nN={vec.shape[1]} | {(vec[-1,:]>.75).sum()}>80% | {(vec[-1,:]<.25).sum()}<25%')

vec_ma = np.array([moving_average(vec[:,i], 50) for i in range(vec.shape[1])])
plt.figure(figsize=(15,4))
plt.plot(vec_ma.transpose(),alpha=.8);
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1);
plt.plot(np.array([val['rsa_vals'] for val in training_logs])[:,:,0].transpose());
plt.title('corr: Z / Ztx')

plt.subplot(2,2,2);
plt.plot(np.array([val['rsa_vals'] for val in training_logs])[:,:,1].transpose());
plt.title('corr: Z / Stx')

plt.subplot(2,2,3);
plt.plot(np.array([val['rsa_vals'] for val in training_logs])[:,:,2].transpose());
plt.title('corr: S / Ztx')

plt.subplot(2,2,4);
plt.plot(np.array([val['rsa_vals'] for val in training_logs])[:,:,3].transpose());
plt.title('corr: S / Stx')

plt.figure(figsize=(15,5))
vec_diff = np.array([val['rsa_vals'] for val in training_logs])[:,:,3].transpose() - np.array([val['rsa_vals'] for val in training_logs])[:,:,2].transpose()
plt.plot(vec_diff);
plt.title('S/Stx - S/Ztx')
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.hist(np.array([val['rsa_vals'] for val in training_logs])[:,:,3].transpose()[-1,:])
plt.title('S/S - S/Z')
plt.title('S')

plt.subplot(1,2,2)
plt.hist(vec_diff[-1,:])
plt.title('S/S - S/Z')
```
